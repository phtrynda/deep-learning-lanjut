```python
import pandas as pd

# Function to generate synthetic sentences
def generate_sentences(sentiment, count):
    sentences = []
    if sentiment == 'positive':
        base_phrases = [
            "Ini sangat bagus.", "Saya sangat senang dengan ini.", "Luar biasa sekali.",
            "Puas sekali.", "Sangat direkomendasikan.", "Sangat berkualitas.",
            "Sempurna.", "Indah sekali.", "Menyenangkan.", "Hebat sekali."
        ]
        for i in range(count):
            sentences.append(f"{base_phrases[i % len(base_phrases)]} Ini adalah pengalaman yang " + ['luar biasa', 'fantastis', 'baik', 'positif'][i % 4] + ".")
    elif sentiment == 'negative':
        base_phrases = [
            "Ini sangat buruk.", "Saya tidak senang dengan ini.", "Mengecewakan sekali.",
            "Tidak puas.", "Tidak direkomendasikan.", "Sangat tidak berkualitas.",
            "Jelek sekali.", "Buruk.", "Menyedihkan.", "Tidak sesuai harapan."
        ]
        for i in range(count):
            sentences.append(f"{base_phrases[i % len(base_phrases)]} Ini adalah pengalaman yang " + ['buruk', 'mengerikan', 'negatif', 'tidak memuaskan'][i % 4] + ".")
    else: # neutral
        base_phrases = [
            "Ini biasa saja.", "Ini tentang produk.", "Tidak ada yang istimewa.",
            "Ini hanya sebuah laporan.", "Hanya fakta.", "Ini adalah barang.",
            "Informasi ini.", "Ini adalah data.", "Sesuai deskripsi.", "Tidak ada komentar."
        ]
        for i in range(count):
            sentences.append(f"{base_phrases[i % len(base_phrases)]} Ini adalah " + ['sebuah pernyataan', 'deskripsi produk', 'observasi', 'data'][i % 4] + ".")
    return sentences

# Generate 100 positive sentences
positive_sentences = generate_sentences('positive', 100)
positive_labels = ['positive'] * 100

# Generate 100 negative sentences
negative_sentences = generate_sentences('negative', 100)
negative_labels = ['negative'] * 100

# Generate 100 neutral sentences
neutral_sentences = generate_sentences('neutral', 100)
neutral_labels = ['neutral'] * 100

# Combine all data
all_sentences = positive_sentences + negative_sentences + neutral_sentences
all_labels = positive_labels + negative_labels + neutral_labels

# Create a DataFrame
training_df = pd.DataFrame({
    'sentence': all_sentences,
    'sentiment': all_labels
})

print(training_df.head())
print(training_df.info())
```

```python
training_df.to_csv('training_data.csv', index=False)
print("DataFrame 'training_df' successfully saved to 'training_data.csv'.")
```

```python
def generate_test_sentences(count):
    sentences = []
    base_phrases = [
        "Informasi yang disajikan cukup jelas.",
        "Produk ini memiliki beberapa fitur.",
        "Cuaca hari ini cukup cerah.",
        "Berita terbaru datang dari ibu kota.",
        "Penelitian ini membahas tentang tren pasar.",
        "Data menunjukkan adanya perubahan kecil.",
        "Sistem ini beroperasi sesuai standar.",
        "Laporan ini mencakup berbagai aspek.",
        "Ini adalah ringkasan dari peristiwa.",
        "Kegiatan tersebut dijadwalkan besok."
    ]
    descriptors = [
        "dengan beberapa detail.",
        "yang perlu dianalisis lebih lanjut.",
        "dan observasi sedang berlangsung.",
        "tanpa kesimpulan yang pasti.",
        "yang relevan untuk studi ini.",
        "dan hasilnya sedang ditinjau.",
        "dalam konteks yang lebih luas.",
        "sebagai bagian dari proyek.",
        "yang belum lengkap.",
        "dan dapat diinterpretasikan."
    ]

    for i in range(count):
        sentences.append(f"{base_phrases[i % len(base_phrases)]} {descriptors[i % len(descriptors)]}")
    return sentences

# Generate 400 test sentences
test_sentences = generate_test_sentences(400)

# Create a DataFrame for test data
test_df = pd.DataFrame({
    'sentence': test_sentences
})

print(test_df.head())
print(test_df.info())
```

```python
import pandas as pd

training_df = pd.read_csv('training_data.csv')
test_df = pd.read_csv('test_data.csv')

print("Training DataFrame loaded successfully:")
print(training_df.head())
print("\nTest DataFrame loaded successfully:")
print(test_df.head())
```

```python
import transformers
from transformers import AutoTokenizer

# Load pre-trained tokenizer
tokenizer = AutoTokenizer.from_pretrained('indobenchmark/indobert-base-p1')

print("Tokenizer loaded successfully:")
print(tokenizer)
```

```python
max_length = 128 # Define a suitable max_length, can be adjusted based on typical sentence length or model requirements

# Tokenize training data
encoded_training_data = tokenizer.batch_encode_plus(
    training_df['sentence'].tolist(),
    add_special_tokens=True,
    return_attention_mask=True,
    padding='max_length',
    truncation=True,
    max_length=max_length,
    return_tensors='pt'
)

train_input_ids = encoded_training_data['input_ids']
train_attention_masks = encoded_training_data['attention_mask']

# Tokenize test data
encoded_test_data = tokenizer.batch_encode_plus(
    test_df['sentence'].tolist(),
    add_special_tokens=True,
    return_attention_mask=True,
    padding='max_length',
    truncation=True,
    max_length=max_length,
    return_tensors='pt'
)

test_input_ids = encoded_test_data['input_ids']
test_attention_masks = encoded_test_data['attention_mask']

print("Training input IDs shape:", train_input_ids.shape)
print("Training attention masks shape:", train_attention_masks.shape)
print("Test input IDs shape:", test_input_ids.shape)
print("Test attention masks shape:", test_attention_masks.shape)
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Fit and transform the 'sentiment' column of training_df
train_labels = label_encoder.fit_transform(training_df['sentiment'])

print("Original sentiment labels:", training_df['sentiment'].unique())
print("Encoded numerical labels:", label_encoder.classes_)
print("First 5 encoded training labels:", train_labels[:5])
print("Shape of training labels:", train_labels.shape)
```

```python
import torch
from torch.utils.data import TensorDataset, DataLoader

print("PyTorch utilities imported successfully.")
```

```python
import numpy as np

# Convert train_labels to a PyTorch tensor with torch.long data type
train_labels_tensor = torch.tensor(train_labels, dtype=torch.long)

# Create a TensorDataset for training data
train_dataset = TensorDataset(train_input_ids, train_attention_masks, train_labels_tensor)

# Create a DataLoader for the training dataset
batch_size = 16
train_dataloader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)

print("Training labels converted to PyTorch tensor.")
print("TensorDataset created for training data.")
print(f"DataLoader created for training data with batch size: {batch_size}.")
print("First batch (input_ids, attention_mask, labels) shapes:")
for batch in train_dataloader:
    print(f"  Input IDs shape: {batch[0].shape}")
    print(f"  Attention Mask shape: {batch[1].shape}")
    print(f"  Labels shape: {batch[2].shape}")
    break
```

```python
from transformers import AutoModelForSequenceClassification

# Get the number of unique labels from the LabelEncoder
num_labels = len(label_encoder.classes_)

# Load the pre-trained model for sequence classification
model = AutoModelForSequenceClassification.from_pretrained(
    'indobenchmark/indobert-base-p1',
    num_labels=num_labels,
    output_attentions=False, # Whether the model returns attentions weights.
    output_hidden_states=False, # Whether the model returns all hidden-states.
)

# Move model to GPU if available
if torch.cuda.is_available():
    model.cuda()

print("Pre-trained Transformer model loaded successfully.")
print(f"Model configured for {num_labels} labels.")
```

```python
from torch.optim import AdamW
from transformers import get_linear_schedule_with_warmup

# Set training parameters
epochs = 3
total_steps = len(train_dataloader) * epochs

# Initialize optimizer and learning rate scheduler
optimizer = AdamW(model.parameters(), lr=2e-5, eps=1e-8)
scheduler = get_linear_schedule_with_warmup(
    optimizer, num_warmup_steps=0, num_training_steps=total_steps
)

print("Optimizer (AdamW) and learning rate scheduler initialized.")
print(f"Total training steps: {total_steps}")
```

```python
import time
import datetime

# Function to format time
def format_time(elapsed):
    elapsed_rounded = int(round(elapsed))
    return str(datetime.timedelta(seconds=elapsed_rounded))

# Set device to GPU if available, else CPU
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

# Store the average loss after each epoch so we can plot them.
avg_train_loss = []

# For each epoch...
for epoch_i in range(epochs):
    print(f'\n======== Epoch {epoch_i + 1} / {epochs} ========\nTraining...')

    t0 = time.time()
    total_train_loss = 0
    model.train() # Set the model to training mode.

    # For each batch of training data...
    for step, batch in enumerate(train_dataloader):
        if step % 50 == 0 and not step == 0:
            elapsed = format_time(time.time() - t0)
            print(f'  Batch {step:>5,}  of  {len(train_dataloader):>5,}.    Elapsed: {elapsed}.')

        b_input_ids = batch[0].to(device)
        b_input_mask = batch[1].to(device)
        b_labels = batch[2].to(device)

        model.zero_grad()

        outputs = model(
            b_input_ids,
            attention_mask=b_input_mask,
            labels=b_labels
        )
        loss = outputs.loss
        total_train_loss += loss.item()

        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0) # Clip the norm of the gradients to 1.0 to prevent the "exploding gradients" problem.
        optimizer.step()
        scheduler.step()

    avg_train_loss.append(total_train_loss / len(train_dataloader))

    print(f'\n  Average training loss: {avg_train_loss[epoch_i]:.2f}')
    print(f'  Training epoch took: {format_time(time.time() - t0)}')

print('\nTraining complete!')
print(f'Average training loss for all epochs: {avg_train_loss}')
```

```python
from torch.utils.data import TensorDataset, DataLoader

# Create a TensorDataset for test data
test_dataset = TensorDataset(test_input_ids, test_attention_masks)

print("TensorDataset created for test data.")
```

```python
batch_size = 16 # Use the same batch_size as training data
test_dataloader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

print(f"DataLoader created for test data with batch size: {batch_size}.")
print("Shuffling for test data is set to False.")
```

```python
model.eval()

print("Model set to evaluation mode.")
```

```python
predictions = []

print("Empty list 'predictions' initialized to store model outputs.")
```

```python
import torch

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Predict
for batch in test_dataloader:
    # Add batch to GPU
    batch = tuple(t.to(device) for t in batch)

    # Unpack the inputs from our dataloader
    b_input_ids, b_input_mask = batch

    # Telling the model not to compute or store gradients, saving memory and speeding up prediction
    with torch.no_grad():
        # Forward pass, calculate logit predictions
        outputs = model(
            b_input_ids,
            attention_mask=b_input_mask
        )

    logits = outputs.logits

    # Move logits and labels to CPU
    predictions.append(logits.detach().cpu())

print("Finished iterating through test_dataloader and collecting logits.")
```

```python
predictions = torch.cat(predictions, dim=0)

print("All collected logits concatenated into a single tensor.")
print("Shape of concatenated predictions:", predictions.shape)
```

```python
predicted_labels = torch.argmax(predictions, dim=1).cpu().numpy()

print("Predicted numerical labels generated.")
print("First 5 predicted numerical labels:", predicted_labels[:5])
print("Shape of predicted numerical labels:", predicted_labels.shape)
```

```python
predicted_sentiment_labels = label_encoder.inverse_transform(predicted_labels)

print("Predicted sentiment labels generated.")
print("First 5 predicted sentiment labels:", predicted_sentiment_labels[:5])
```
